```python
import numpy as np

samples = 1000
dimension = 4

beta = np.random.randn(dimension)

def S(beta, gamma):
    if gamma >= beta:
        return 0
    else:
        return beta - gamma if beta > 0 else beta + gamma

gamma = 0.5
y_tilde = np.zeros((dimension,))
X = np.random.randn(samples, dimension)
y = np.random.randn(samples, 1)
y = (X[:, 0] + X[:, 2] + X[:, 3]).reshape(samples, 1)
y -= np.mean(y)
y /= np.std(y)

for _ in range(50):
    for j in range(dimension):
        # Calculate y_tilde
        y_tilde = np.zeros((samples,))
        for i in range(samples):
            for k in range(dimension):
                y_tilde[i] = X[i, k] * beta[k]
        f1 = beta[j]
        for i in range(samples):
            f1 += X[i, j] * (y[i] - y_tilde[i])
        beta[j] = S(f1 / samples, gamma)
    print(beta)
```
